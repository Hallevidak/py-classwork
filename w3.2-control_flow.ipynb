```python
# If statements 

condition = "" # everything in python is truthy unless it is one of the defined falsy values above 

if condition: 
    pass # does nothing, lets us continue logic without error 
    print("condition is true")
elif bool(condition) == False: #condition is an empty string unless add boolean (elif bool(condition)) - difference between determining 
    # if string is false vs checking if string is equal to false
    pass 
    print("condition is false")
else: 
    print("I got to the else statement")
```

```python
# If statements 

condition = "" # everything in python is truthy unless it is one of the defined falsy values above 

if condition: 
    pass # does nothing, lets us continue logic without error 
    print("condition is true")
elif not condition: #dont need pass - just lets continue logic without error 
    print("condition is false")
else: 
    print("I got to the else statement")
    
print (f"not statement = {not condition}") # reverses the truthy/falsy value of the condition 
```

```python
# Let's make this better with try except 

num = input("Choose a number between 1 and 10:")

try: # allows the code to not shut down if there is an error
    num = int(num)
    if num < 3: 
        print(f"Number is less than 3: {num}")
    elif num > 6: 
        print(f"Number is greater than 6: {num}")
    else: 
        print(f"num is between 3 and 6: {num}")
except Exception as e: 
    print(f"Well that didn't work, try again: {e}")
    
# can except: a type error, value error, etc. 
```

```python
# Nested if statements 
# type hinting: str, float, int, bool

variable: str = input(f"Please enter a value:")

if variable.isnumeric():
    try:
        num = float = float(variable) #can't use a string to compare to a number 
        if num > 10:
            print(f"The value is too high: {num}, please try again")
        else: 
            print(f"{num} is a good choice")
    except Exception as e: 
        print(f"There was an error: {e}")
elif variable.isalpha():
    pass 
else: 
    print("Please enter either a number or a letter")
```
