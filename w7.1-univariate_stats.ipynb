```python
import pandas as pd 

url = 'https://raw.githubusercontent.com/nkmwicz/data-for-students/refs/heads/main/soccer-players.csv'

df = pd.read_csv(url)
df.head()
```

```python
def convert_height(cm) -> float:
    '''
    Convert height from centimeters to feet.
    
    ARGS:
    cm(int): the num of centimeters 
    RETURNS:
    feet (float): the num of feet 
    '''
    return (cm * 0.393701) / 12
```

```python
df['height_in_ft'] = df['height_in_cm'].apply(convert_height)
df['height_in_ft']
```

```python
def eda(df):
    '''
    Evaluate basic statistics on a dataframe
    
    ARGS:  
        df (pandas.DataFrame): the dataframe to be evaluated
    Returns:
        None
    '''
    print(f"('='*5)DF Shape: {df.shape}{'='*5}")
    num_cols = []
    cat_cols = []
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            num_cols.append(df[col])
        else: 
            cat_cols.append(df[col])
    print(f"{'*'*3}Numerical Columns{'*'*3}")
    for col in num_cols:
        print(col)
        print(f"\tMin: {df[col].min()}")
        print (f"\tMax: {df[col].max()}")
```

```python
eda(df)
```
