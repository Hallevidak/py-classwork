```python
import pandas as pd

df = pd.DataFrame(data = {'Name': ['Nate', 'John', 'Jane', 'Mandy'], 
                        'Age': ['39', '23', '31', '25'], 
                        'Sex': ['M', 'M', 'F', 'F'], 
                        'Email': ['nate@gmail.com', 'john@gmail.com', 'jane@gmail.com', 'mandy@gmail.com'], 
                        'Purchases': ['34','12', '42', '8']
                        })

print (df)
```

```python
df = df.rename(columns={'Sex' :'Gender'})
```

```python
print(df)
```

```python
df['Active'] = True
df
```

```python
df['Purchases'] = pd.to_numeric(df['Purchases']) 
```

```python
df['Purchases'] = df['Purchases'] + 5
df
```

```python
df.loc[df['Purchases'] > 20]
```

```python
for item in df.index:
    if df.at[item, 'Gender'] == 'M': 
        df.at[item, 'Gender'] = 'Male'
    if df.at[item, 'Gender'] == 'F': 
        df.at[item, 'Gender'] = 'Female'
print(df)
```
