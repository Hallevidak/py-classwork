```python
# set 
# no duplicates, not indexed (no order), iterable (can loop through it - 
# do something repetively), collection, mutable (can change it - add/remove from it)

set1 = {"Ben", "Jason", "Darlene", "Jason", "Erin"}  # curly braces
set_declared = set(("Ben", "Jason", "G", "G", "Darlene", "Jason"))
# declaring sets, a collection in parentheses is a tuple that we turned into a set 
print(f"set1 = {set1}")
print(f"set_declared = {set_declared}")
```

```python
# method to find overlaps 
overlaps = set1.intersection(set_declared)
overlaps
```

```python
# method to find difference between two sets 
diff = set1.difference(set_declared)
diff #last item in a cell will be read in a jupiter notebook
```

```python
# method to find difference between two sets 
diff = set_declared.difference(set1)
diff #last item in a cell will be read in a jupiter notebook
```

```python
set1.add("Halle")
set1 #add/update a name to the set 
```

```python
set1.pop("Halle")
set1 #ignore this 
```

```python
set1.remove("Darlene")
set1 #delete a name from the set 
```

```python
# tuple 
# nonmutable (cannot change it - add/remove from it), indexed (ordered), 
# iterable (can repeat),collection

tuple1 = (1,2,3,4)
tuple_declared = tuple((3,4,5,6))

print(f"tuple1 = {tuple1}")
print(f"tuple_declared = {tuple_declared}")
```

```python
tuple1 = (1,2,3,4,5) # overwriting not changing 
tuple1
```

```python
tuple1.count(2) # count occurances of a value
```

```python
tuple1.index(1) #return index of value (1 is at the zero value)
```

```python
tuple1[0] # return value at the index 0
```

```python
print(f"tuple1 = {tuple1}")
tuple1[0] #tuples dont change, returning bank info or series of items 
# that were purchased, we want it to not be changable
```

```python
#list 
#mutable (can change), duplicated(can have duplicates), indexed [index notation], 
# iterable, string has an index because it is a collection of characters that needs an order 
#if they are ordered, they have an index-index just indicates what order something exists in

#can add to a list, and a set, but not a tuple 

list1 = ["G", "G", "Paola", "Emil"]
list_declared = list(("G", "G", "Paola", "Emil"))

print(f"list1 = {list1}")
print(f"list_declared = {list_declared}")
```

```python
print(f"list1 = {list1}")
print(f"first index = {list1[0]}")
print(f"last item in list1 = {list[-1]}")
print(f"series between 2 & 3 items = {list1[1:2]}") #including 1, up to 2 (2 is exclusive))
```

```python
print(f"series between 2 & 3 items = {list1[1:3]}") #including 1, up to 3 (3 is exclusive)
```

```python
print(f"everything before index 3 = {list1[:3]}") #everything before index 3 (3 is exclusive))
```

```python
print(f"everything after the first = {list1[1:]}') #everything after index 0 (1 is inclusive))")
```

```python
print(f"last item in list1 = {list1[-1]}")
```

```python
#list methods 
list1.append("Claire")#add to a list
list2 = list1.append("Nathan") #append does not create a new list, it just changes the existing list
print(f"appended Claire = {list1}")

removed = list1.pop() #remove (defaults to last item in index), it returns the removed item,
# changed the list in place
#change index
#combine lists 

print(f"appended Claire = {list1}")
print(f"removed {removed} . New list is {list1}")

removed = list1.pop()
```

```python
list1[1] = "J" # change index using index notation, change value at index 1 to J
print(f"index 1 of list 1 changed to J = {list1}")

combined_list1 = [list1, list_declared] #combine lists and lists 
print(f"combined list = {combined_list}")

combined_lists = [*list1, *list_declared] #combine lists into one list #star notation to rip items 
# out of list and into one list that combines them both 
print(f"comibined lists into one list = {combined_list}")

list1.extend(list_declared) #combine lists into one list 
print(f"list1 after extension = {list1}")
```
