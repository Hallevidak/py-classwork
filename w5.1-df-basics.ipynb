DF can have: 
- column labels (each series is going to be a column, x-axis)
- index labels (y-axis)

2 ways to create a DF: 
- dictionary creation 
- row-based creation 

Rules: 
- columns need to be homogenous (same data type) (cannot mix strings and numbers, etc. in one column) (all have to be strings, or whatever type)
- every column has to have the same number of data entries (every series must have the same amount) 

```python
#DICTIONARY CREATION 

data = {
    'name': ['Paige', 'Alexis', 'Sarah', 'Halle'],
    'age': ['33', '27', '25', '21'],
    'favorite color': ['green', 'orange', 'green', 'blue']
}

students_df = pd.DataFrame(data=data)
students_df = students_df.set_index('name')
students_df #changes index to the name
```

```python
students_df = students_df.reset_index()
students_df
```

```python
#ROW-BASED CREATION

import random 

names = ['Messi', 'Ronaldo', 'Neymar', 'Pulisc', 'Mbappe']
goals_per_game = [random.random() for _ in names]
games = [round(random.random() * 1000) for _ in names] # * 1000, 1) declares how many decimals to round to 
games

cols = ['name', 'goals per game', 'games']
stats = [[names[i], goals_per_game[i], games[i]] for i in range(5)]

stats_df = pd.DataFrame(data=stats, columns=cols) #changes the list of lists into a data frame table 
stats_df
```

```python
stats_df['new column'] = 0 
stats_df
```

```python
stats_df = stats_df.drop(columns=['new column'])
stats_df
```

```python
stats_df['goals'] = stats_df['goals per game'] * stats_df['games']
stats_df
```

```python
stats_df = stats_df.rename(columns={'goals per game': 'goals_p_game'})
# stats_df.rename9(columns={'goals per game': 'goals_p_game'}, inplace = True)
stats_df
```

```python
#call the column 
stats_df['games']
```

```python
stats_df['name'] = stats_df['name'].str.lower()
stats_df
```

```python
stats_df.loc[1:2, 'name'] # in df, second items after colon is inclusive 
```

```python
stats_df.loc[:, 'name':'goals']
```

```python
# BOOLEAN LOGIC WITH LOC 

stats_df.loc[stats_df['goals_p_game'] > 0.5] #rows go first 
```

```python
stats_df.loc[stats_df['goals_p_game'] > 0.5, 'name']
```

```python
stats_df.loc[stats_df['name'] == 'messi', 'name'] = 'ronaldo'
stats_df.loc[stats_df['goals_p_game'] > 0.5, 'name'] = 'messi'
stats_df
```
