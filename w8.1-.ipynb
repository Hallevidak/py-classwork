```python
import scipy.stats as stats 
```

```python
# Histogram in SNS 
df1 = df.loc[df['height_in_cm']>150]
plt.figure(figsize=(5,3))
sns.set_theme(style = 'darkgrid')
hist = sns.histplot(data=df1, x='height_in_cm', kde=True,) #kernal of density estimate - the dark blue line 
plt.xlabel('Height (cm)')
hts = df1['height_in_cm']
kurt = stats.kurtosis(hts, nan_policy='omit')
sk = stats.skew(hts, nan_policy = 'omit')
text = f'Skew: {sk:.3f}\nKurtosis: {kurt:.3f}'
# x=0.2, y=0.95,  s=text,  ha='center',  va='center',  fontsize=12,  color='hotpink'
plt.text(
    x=0.25, #indicates percentage of the x-axis, 0.5 is the middle of the x-axis
    y=0.9, 
    s=text, 
    ha='center', 
    va='center', 
    fontsize=12, 
    color='hotpink',
    transform=plt.gca().transAxes) 
plt.axvline(x=hts.mean(), color='purple', linestyle='--', label = 'Mean') #(line Vertically across the middle of the plot)
plt.axvline(x=hts.median(),color='cyan', linestyle='--', label = 'Median') #(line Horizontally across the middle of the plot)
plt.legend()
plt.show()
```
