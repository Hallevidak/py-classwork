WHILE LOOP

```python
count = 0
while True: 
    if count >= 10:
        break #breaks the loop 
    count += 1 #count =count + 1 (sugaring the code, making it easier)
    print(count)
```

```python
# a better option 
count = 0 

while count <= 10: 
    print(count) 
    count += 1 
```

```python
is_pass_complete: bool = False #starting with "is" is always a boolean 

while not is_pass_complete: # not reverses it (true)
    password = input("Enter Password in only alphetic characters.")
    if not password.isalpha():
        print("Please choose another password with only alphetic characters.")
        continue # restarts or continues the loop from here. Reads nothing below 
    is_pass_complete = True 
    print("Password Set")
```

FOR LOOP

```python
# for loop 

from typing import List 

nums: List[int] = [1,2,3,4,5]
# can name this whatever you want (item, num, etc. - temporary variable used to call 
# each item in the collection in the code block
for num in nums:
    num += 1
    if num % 3 == 0: 
    # if num % 2 != 0: if odd
        # break would break the loop
        continue #dont print anything divisible by 3 
    print(num)
    
```

NESTED LOOP

```python
sentence = "This is a Python course."

words: List[str] = sentence.split(" ")
used_vowels = []
#whatever is delcared in loop only exists in loop, must declare outside otherwise

for word in words: 
    vowels = ["a", "e", "i", "o", "u", "y"]
    for letter in word: 
            if letter in vowels: 
                used_vowels.append(letter)
       

print(used_vowels)
```

FUNCTIONS 

```python
# what are functions? 
# function take an input (argument), processes it, and return 
# an output (return)

# def defines a function, give it a name, ():, #what comes after the parenthese and before the colon is just type-hinting 
def shorten_names(names_list: List[str], chars: int = 3) -> List[str]:
    """ #doc (document) string 
    Returns a list of strings that are the first three
    characters of the word within a period appended.
    
    Args: 
        names_list 9List[str]): A list of strings that will be shortened. 
        chars (int): The number of characters to shorten each string to. Defaults to 3
    Return: 
        List[str]: A list of shortened strings to 3 characters 
    """
    if type(chars) != int:
        raise TypeError("Chars must be an int")
    if chars < 0: 
        raise ValueError("Chars must be greater than 0")
    names_list = [f"{name[:chars]}." for name in names_list]
    return names_list
```

```python
bar_names = ["Ottoman Empire", "France", "Spain"]
shorten_names(bar_names, chars=4)
```
